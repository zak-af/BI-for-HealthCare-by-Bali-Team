```python
import pandas as pd
```

```python
#import data
indeks_ = pd.read_csv('indeks_kesehatan.csv')
keluhan_ = pd.read_csv('keluhan_kesehatan.csv')
ims_ = pd.read_csv('jml_ims.csv')
balita_ = pd.read_csv('jml_balita.csv')
```

```python
# copy data
data1 = indeks_.copy()
data2 = keluhan_.copy()
data3 = ims_.copy()
data4 = balita_.copy()
```

```python
data4.info()
```

```python
#drop column
data1 = data1.drop(['id','kode_provinsi', 'satuan'], axis=1)
data2 =data2.drop(['id', 'kode_provinsi', 'satuan'], axis=1)
data3 =data3.drop(['id', 'kode_provinsi', 'satuan'], axis=1)
data4 =data4.drop(['id', 'kode_provinsi', 'satuan'], axis=1)

#merge
df = data1.merge(data2, on=('kode_kabupaten_kota', 'nama_provinsi', 'nama_kabupaten_kota', 'tahun'), how='outer')
df = df.merge(data3, on=('kode_kabupaten_kota', 'nama_provinsi', 'nama_kabupaten_kota', 'tahun'), how='outer')
df = df.merge(data4, on=('kode_kabupaten_kota', 'nama_provinsi', 'nama_kabupaten_kota', 'tahun'), how='outer')
```

```python
df
```

```python
 #ubah kolomn
df = df[['tahun', 'nama_provinsi', 'kode_kabupaten_kota', 'nama_kabupaten_kota', 'indeks_kesehatan', 'persentase_penduduk', 'jumlah_kasus', 'jumlah_balita']]
```

```python
df.info()
```

```python
# hapus tahun

df1 = df.copy()

clean_tahun = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
df1 = df1[df1.tahun.isin(clean_tahun) == False]
```

```python
df1.isna().sum()
```

```python
#rename colomns
df1.rename(columns = {'jumlah_kasus': 'jml_ims', 'jumlah_balita': 'jml_balita'}, inplace=True)
```

```python
df1.isna().sum()
```

```python
#fillna kabupaten banjar
df1 = df1.fillna(78.14)
```

```python
df1.to_csv('jabar_sehat20_21.csv')
```
