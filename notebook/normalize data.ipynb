```python
#normalize data

import pandas as pd
from sklearn.preprocessing import normalize
```

```python
data1 = pd.read_csv('jabar_sehat.csv')
data2 = pd.read_csv('jabar_sehat2.csv')
data3 = pd.read_csv('kdrt_2018_2021.csv')
data4 = pd.read_csv('kekerasan_pendidikan_2018_2021.csv')
data5 = pd.read_csv('kekerasan_usia_2018_2021.csv')
```

```python
norm_data1 = data1.copy()
norm_data2 = data2.copy()
norm_data3 = data3.copy()
norm_data4 = data4.copy()
norm_data5 = data5.copy()
```

```python
norm_data5
```

```python
kolom3 = ['jml_kdrt']
```

```python
kolom4 = ['jml_kelompok_pendidikan']
kolom5 = ['jml_kelompok_usia']
```

```python
#normalize 345
norm_data3[kolom3] = normalize(X=norm_data3[kolom3], norm='l2', axis=1)
norm_data4[kolom4] = normalize(X=norm_data4[kolom4], norm='l2', axis=1)
norm_data5[kolom5] = normalize(X=norm_data5[kolom5], norm='l2', axis=1)
```
