```python
import pandas as pd
```

```python
data1 = pd.read_csv('jabar_sehat20_21.csv')
data2 = pd.read_csv('kdrt_2018_2021.csv')
data3 = pd.read_csv('kekerasan_pendidikan_2018_2021.csv')
data4 = pd.read_csv('kekerasan_usia_2018_2021.csv')
```

```python
data1.head()
```

```python
# wrangling data1
data1 = data1.drop(['Unnamed: 0', 'indeks_kesehatan', 'persentase_penduduk'], axis=1)

#drop tahun
data1.drop(data1[data1['tahun']== 2020].index, inplace= True)
```

```python
#wrangling data 2, 3, 4

data2_group = data2.copy()
data3_group = data3.copy()
data4_group = data4.copy()

# drop tahun 2020
data2_group.drop(data2_group[data2_group['tahun']== 2018].index, inplace= True)
data3_group.drop(data3_group[data3_group['tahun']== 2018].index, inplace= True)
data4_group.drop(data4_group[data4_group['tahun']== 2018].index, inplace= True)
```

```python
#group by berdasarkan nama kabupaten
data2_ = data2_group.groupby(['nama_kabupaten_kota'])[['jml_kdrt']].sum()
data3_ = data3_group.groupby(['nama_kabupaten_kota'])[['jml_kelompok_pendidikan']].sum()
data4_ = data4_group.groupby(['nama_kabupaten_kota'])[['jml_kelompok_usia']].sum()
```

```python
data2_.head(3)
```

```python
# merge data
data = data1.merge(data2_, on=('nama_kabupaten_kota'), how='outer')
data = data.merge(data3_, on=('nama_kabupaten_kota'), how='outer')
data = data.merge(data4_, on=('nama_kabupaten_kota'), how='outer')

```

```python
score = data.copy()
```

```python
score.head(1)
```

```python
#membuat rank


score['mean'] = score.mean(axis=1)
score['total'] = score['jml_ims'] + score['jml_balita'] + score['jml_kdrt'] + score['jml_kelompok_pendidikan'] + score['jml_kelompok_usia']
score['ranking'] = score['mean'].rank(method='min')
```

```python
score.to_csv('data_score.csv')
```
